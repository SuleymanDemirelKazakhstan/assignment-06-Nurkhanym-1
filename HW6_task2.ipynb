```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\NRS\\Desktop\\Jupyter\\IMDB Dataset.csv')
df
```

```python
df1 = df[0:100]
```

```python
x = df1['review'].str.lower()
y = df1['sentiment']
x.str.replace("\\", "");
x.str.replace("-", "");

d ={}
for i in range(0, len(x)):
    doc1 = nlp(x[i])
    for token in doc1:
        if(token.tag_ == 'JJ' and (token.lemma_ not in d) and y[i] == 'negative'):
            d[token.lemma_] = 1, 0
        elif(token.tag_ == 'JJ' and (token.lemma_ not in d) and y[i] == 'positive'):
            d[token.lemma_] = 0, 1
        elif(token.tag_ == 'JJ' and (token.lemma_ in d) and y[i] == 'positive'):
            d[token.lemma_] = d[token.lemma_][0], d[token.lemma_][1]+1
        elif(token.tag_ == 'JJ' and (token.lemma_ in d) and y[i] == 'negative'):
            d[token.lemma_] = d[token.lemma_][0]+ 1, d[token.lemma_][1]
                
            
```

```python
d
```

```python
df2 = pd.DataFrame(d.items(), columns=['word', 'sent'])
```

```python
df3 = pd.DataFrame(df2['sent'].to_list(), columns=['negative', 'positive'])
```

```python
df3 = pd.concat([df2, df3],axis=1, ignore_index=True, sort=False)
df3 = df3.drop(1, axis=1)
```

```python
df3.rename(columns={0: 'word', 2: 'negative', 3: 'positive'}, inplace=True)
```
