```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\NRS\\anaconda3\\bbc-news-data.csv', sep='\t')
df = df.drop('filename', axis=1)
df = df.drop('title', axis=1)
df
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
x = df['content']
y = df['category']

d ={}
for i in range(0, len(x)):
    doc = nlp(x[i])
    for ent in doc.ents:
        if(ent.text not in d and y[i] == 'business'):
            d[ent.text] = 1, 0, 0, 0, 0
        elif(ent.text not in d and y[i] == 'entertainment'):
            d[ent.text] = 0, 1, 0, 0, 0
        elif(ent.text not in d and y[i] == 'politics'):
            d[ent.text] = 0, 0, 1, 0, 0
        elif(ent.text not in d and y[i] == 'sports'):
            d[ent.text] = 0, 0, 0, 1, 0
        elif(ent.text not in d and y[i] == 'tech'):
            d[ent.text] = 0, 0, 0, 0, 1
        elif(ent.text in d and y[i] == 'tech'):
            d[ent.text] = d[ent.text][0], d[ent.text][1], d[ent.text][2], d[ent.text][3], d[ent.text][4]+1 
        elif(ent.text in d and y[i] == 'sports'):
            d[ent.text] = d[ent.text][0], d[ent.text][1], d[ent.text][2], d[ent.text][3]+1, d[ent.text][4]
        elif(ent.text in d and y[i] == 'politics'):
            d[ent.text] = d[ent.text][0], d[ent.text][1], d[ent.text][2]+1, d[ent.text][3], d[ent.text][4]
        elif(ent.text in d and y[i] == 'entertainment'):
            d[ent.text] = d[ent.text][0], d[ent.text][1]+1, d[ent.text][2], d[ent.text][3], d[ent.text][4]
        elif(ent.text in d and y[i] == 'business'):
            d[ent.text] = d[ent.text][0]+1, d[ent.text][1], d[ent.text][2], d[ent.text][3], d[ent.text][4] 
        
        
              
```

```python
d
```

```python
df1 = pd.DataFrame(d.items(), columns=['word', 'topic'])
```

```python
df2 = pd.DataFrame(df1['topic'].to_list(), columns=['business', 'entertainment','politics','sport','tech'])
```

```python
df2
```

```python
df2 = pd.concat([df1, df2],axis=1, ignore_index=True, sort=False)
df2 = df2.drop(1, axis=1)
```

```python
df2.rename(columns={0: 'word', 2: 'business', 3: 'entertainment', 4: 'politics', 5: 'sport', 6: 'tech'}, inplace=True)
```
